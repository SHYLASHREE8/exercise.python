```python
#palindrome
#concepts: functions,conditions
def is_palindrome(text):
  text=text.lower().replace(" ","")
  return text == text[: : -1]

user_input = input("Enter a word: ")
if is_palindrome(user_input):
  print("It is a palindrome")
else:
  print("It is not a palindrome")
```

```python
#simple calculator
#concepts: functions, conditions
def calculator(a,b,operation):
  if operation == '+':
    return a+b
  elif operation == '-':
    return a-b
  elif operation == '*':
    return a*b
  elif operation == '/':
    return a/b
  else:
    return "Invalid operation"

x=float(input("Enter a first num:"))
y=float(input("Enter a second num:"))
operation=input("Choose an operation(+,-,*,/): ")
result=calculator(x,y,operation)
print("Result: ",result)
```

```python
#factorial
#concept: loops, conditions
def factorial(n):
  if n<0:
    return "Negative values are not allowed"
  elif n==0:
    return 1
  else:
    result=1
    for i in range(1, n+1):
      result *= i
    return result
num=int(input("Enter a number: "))
result=factorial(num)
print(f"factorial of {num} is {result}")
```

```python
#finding max number
#concept: lists, loops, comparison
numbers=[10,5,34,68,71]
max_num=numbers[0]
for num in numbers:
  if num>max_num:
    max_num=num
print("Max number is: ",max_num)
```

```python
#word freq counter
#concepts: file i/o, dictionaries, string manipulation
import string
def count_words(filename):
  with open(filename,'r')as file:
    text=file.read().lower()
  for punct in string.punctuation:
    text=text.replace(punct," ")
  words=text.split()
  freq={}
  for word in words:
      freq[word]=freq.get(word,0)+1
  return freq
  # Create a dummy file for testing
with open("sample.text", "w") as f:
    f.write("This is a sample text file.\nThis file is for testing the word frequency counter.\n")
word_counts=count_words("sample.text")
for word, count in word_counts.items():
  print(f"{word}: {count}")
```

```python
#fibonacci
#concepts: recursion, functions, design
def fibonacci(n):
  if n<0:
    return "Invalid input"
  if n==0:
    return 0
  if n==1:
    return 1
  else:
    return fibonacci(n-1)+fibonacci(n-2)
num=int(input("Enter a number: "))
result=fibonacci(num)
print(f"fibonacci of {num} is {result}")
```

```python
#anagarm grouping
#concepts: dictionaries, collections, loops
from collections import defaultdict
def group_anagrams(strs):
  anagram_groups=defaultdict(list)
  for word in strs:
    sorted_word=''.join(sorted(word))
    anagram_groups[sorted_word].append(word)
  return list(anagram_groups.values())
word_list=[["eat", "tea", "ate"], ["tan", "nat"], ["bat"], ["dormitory", "dirtyroom"], ["listen", "silent"], ["triangle", "integral"]]
group_list=group_anagrams(word_list)
print(group_list)
```
